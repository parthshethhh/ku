```python
import pandas as pd
```

```python
df=pd.read_csv('/content/TimeSeries.csv')
df_test = df.copy()
```

```python
#applying normalization
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
df_test_scaled=scaling.fit_transform(df_test[['v1','v2','v3','v4','v5','v6','v7','v8','v9','v10','v11']])
```

```python
df_test_scaled
```

```python
#converting the normalized numpy array tp data frame
df_test=pd.DataFrame(df_test_scaled, columns = df_test.columns)
df_test.head()
```

```python
#checking number of duplicated rows
df_test.duplicated().sum()
```

```python
#displaying duplicated rows
df_test.loc[df_test.duplicated(keep=False),:]
```

```python
df_test=df_test.drop_duplicates(keep='first')
```

```python
df_test.shape
```

```python
#dimensionality reduction using pca
from sklearn.decomposition import PCA
pca=PCA(n_components=2) #number of features choosen as 2
pca.fit(df_test)
```

```python
x_pca=pca.transform(df_test)
```

```python
x_pca.shape
```
